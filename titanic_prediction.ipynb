```python
# Check if column has NaN or Null
nan_cols_test = []
nan_cols_train = []

for col in test_data.columns:
    if test_data[col].isnull().any():
        nan_cols_test.append(col)
        
for col in train_data.columns:
    if train_data[col].isnull().any():
        nan_cols_train.append(col)        

print('Test:',nan_cols_test)
print('Train:',nan_cols_train)
```

```python
# change sex to 0 or 1
def age_binary(sex):
    if sex == 'male':
        return 1
    else:
        return 0
    
train_data['Male'] = train_data['Sex'].apply(age_binary)
train_data[['Sex','Male']]
```
