```python
# Libraries we will prob use
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('./train.csv') # Creates dataframe object from CSV
train_data.tail() # shows end of dataframe
```

```python
test_data = pd.read_csv('./test.csv')
test_data.tail()
```

```python
# Check if column has NaN or Null
nan_cols_test = []
nan_cols_train = []

for col in test_data.columns:
    if test_data[col].isnull().any():
        nan_cols_test.append(col)
        
for col in train_data.columns:
    if train_data[col].isnull().any():
        nan_cols_train.append(col)        

print('Test:',nan_cols_test)
print('Train:',nan_cols_train)
```

```python
# change sex to 0 or 1
def age_binary(sex):
    if sex == 'male':
        return 1
    else:
        return 0
    
train_data['Male'] = train_data['Sex'].apply(age_binary)
train_data[['Sex','Male']]
```

```python
men = train_data.loc[train_data.Sex == 'male']["Survived"]
print(sum(men)/len(men))
pd.get_dummies(train_data['Age'])
```

```python
corr = train_data['Survived'].corr(train_data['Fare'])
corr
```

```python
from sklearn.model_selection import train_test_split

features = train_data[['Age', 'Male', 'Pclass', 'Parch', 'SibSp', 'Fare']]
target = train_data['Survived']

# print(features)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=0)

print('train', type(X_train))
print('test', type(X_test))
```

```python
import xgboost as xgb

train = xgb.DMatrix(X_train, label=y_train)
test = xgb.DMatrix(X_test, label=y_test)
```

```python
# All default params for xgb model EXCEPT objective
param = {
    'max_depth' : 6,
    'learning_rate' : 0.3,
    'objective' : 'binary:hinge',
}
epochs = 100
```

```python
model = xgb.train(param, train)
```

```python
predictions = model.predict(test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```
